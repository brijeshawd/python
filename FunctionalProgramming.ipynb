```python
def solve(A,i=0,B=[]):
  if i <len(A)-1:
    if (A[i]+A[i+1])%2==1:
      if A[i]%2==0:
        B=A.pop(i)
        return solve(A,i,B)
      else:
        B=A.pop(i+1)
        return solve(A,i,B)
    else:
      i+=1
      return solve(A,i,B)
  else:
    return len(B)
```
